```python
# extracting data from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip') 
zip_ref.extractall() # extract file to dir
zip_ref.close() # close file
```

```python
import pandas as pd 
df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head(20)
```

```python
df['Ship Mode'].unique()
```

```python
#rename the column name for best practice
#df.rename(columns={'Order Id':'order_id', 'City':'city'}) manual effort take more time 
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ", "_")
```

```python
df.head()
```

```python
#derive new columns for discont
df['discount']=(df['list_price']*df['discount_percent'])/100
df['sales_price']=df['list_price']-df['discount']
df['profit']=df['sales_price']-df['cost_price']
```

```python
df.head()
```

```python
#drop columns whic are not needed anymore
df.drop(columns=['cost_price','list_price','discount_percent'], inplace=True)
```

```python
#convert order date to date
df.dtypes
```

```python
import pandas as pd
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```
